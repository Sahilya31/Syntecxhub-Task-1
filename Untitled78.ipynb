```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# =====================================================================
# STEP 1: CLEAN AND PREPARE DATA (10,000 Records)
# =====================================================================
np.random.seed(42)
data_size = 10000  # Data size 10,000 kiya gaya hai

sample_data = {
    'InvoiceNo': np.random.randint(500000, 520000, size=data_size),
    'StockCode': np.random.randint(20000, 30000, size=data_size),
    'Quantity': np.random.randint(-10, 100, size=data_size), # Negative values cleaning test ke liye
    'InvoiceDate': pd.date_range(start='2025-01-01', end='2026-06-15', periods=data_size),
    'UnitPrice': np.random.uniform(1.0, 200.0, size=data_size),
    'CustomerID': np.random.choice([12345 + i for i in range(100)] + [np.nan], size=data_size, p=[0.0095]*100 + [0.05])
}
df = pd.DataFrame(sample_data)
print(f"Original Data Shape: {df.shape}")

# 1. Missing CustomerID remove karna
df = df.dropna(subset=['CustomerID'])

# 2. Duplicate entries remove karna
df = df.drop_duplicates()

# 3. Invalid records handle karna (Quantity & Price > 0) - Total Clean Data
df = df[(df['Quantity'] > 0) & (df['UnitPrice'] > 0)]

# 4. Total Spend column create karna
df['TotalSum'] = df['Quantity'] * df['UnitPrice']
print(f"Totally Cleaned Data Shape: {df.shape}\n")


# =====================================================================
# STEP 2: CALCULATE RFM METRICS
# =====================================================================
snapshot_date = df['InvoiceDate'].max() + pd.Timedelta(days=1)

rfm = df.groupby('CustomerID').agg({
    'InvoiceDate': lambda x: (snapshot_date - x.max()).days, # Recency
    'InvoiceNo': 'nunique',                                 # Frequency
    'TotalSum': 'sum'                                       # Monetary
})

rfm.rename(columns={
    'InvoiceDate': 'Recency',
    'InvoiceNo': 'Frequency',
    'TotalSum': 'Monetary'
}, inplace=True)


# =====================================================================
# STEP 3: SEGMENT CUSTOMERS USING STABLE RANKING
# =====================================================================
r_labels = range(4, 0, -1)
f_labels = range(1, 5)
m_labels = range(1, 5)

# rank(method='first') lagane se "Bin edges must be unique" error kabhi nahi aayega
rfm['R'] = pd.qcut(rfm['Recency'].rank(method='first'), q=4, labels=r_labels)
rfm['F'] = pd.qcut(rfm['Frequency'].rank(method='first'), q=4, labels=f_labels)
rfm['M'] = pd.qcut(rfm['Monetary'].rank(method='first'), q=4, labels=m_labels)

rfm['RFM_Segment'] = rfm['R'].astype(str) + rfm['F'].astype(str) + rfm['M'].astype(str)

def segment_customer(row):
    if row['RFM_Segment'] == '444':
        return 'Champions'
    elif row['F'] == 4 and row['M'] == 4:
        return 'Loyal Customers'
    elif row['R'] == 4 and row['F'] == 1:
        return 'New Customers'
    elif row['R'] == 1:
        return 'Churn Risk'
    elif row['R'] == 2 and row['F'] <= 2:
        return 'About to Sleep'
    else:
        return 'Average / Potential Loyalists'

rfm['Customer_Segment'] = rfm.apply(segment_customer, axis=1)


# =====================================================================
# STEP 4: VISUALIZATION (PIE CHART, BAR GRAPH & MORE)
# =====================================================================
sns.set_theme(style="whitegrid")
segment_counts = rfm['Customer_Segment'].value_counts()

# --- 1. PIE CHART ---
plt.figure(figsize=(8, 8))
plt.pie(
    segment_counts,
    labels=segment_counts.index,
    autopct='%1.1f%%',
    startangle=140,
    colors=sns.color_palette('viridis', len(segment_counts)),
    wedgeprops={'edgecolor': 'white', 'linewidth': 2}
)
plt.title('Customer Segments Share (Pie Chart)', fontsize=16, weight='bold')
plt.tight_layout()
plt.show()

# --- 2. BAR CHART (COUNTPLOT) ---
plt.figure(figsize=(12, 5))
sns.barplot(
    x=segment_counts.index,
    y=segment_counts.values,
    palette='viridis',
    hue=segment_counts.index,
    legend=False
)
plt.title('Number of Customers per Segment (Bar Chart)', fontsize=16, weight='bold')
plt.xlabel('Customer Segments', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=25, ha='right')
plt.tight_layout()
plt.show()

# --- 3. ETC: RFM METRICS DISTRIBUTION ---
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
fig.suptitle('Distribution of RFM Metrics (Overall Behavior)', fontsize=16, weight='bold')

sns.histplot(rfm['Recency'], kde=True, ax=axes[0], color='blue')
axes[0].set_title('Recency Distribution (Days)')

sns.histplot(rfm['Frequency'], kde=True, ax=axes[1], color='green')
axes[1].set_title('Frequency Distribution (Orders)')

sns.histplot(rfm['Monetary'], kde=True, ax=axes[2], color='purple')
axes[2].set_title('Monetary Distribution (Total Spend)')

plt.tight_layout()
plt.show()

# Final output save karna
rfm.to_csv('Syntecxhub_10k_Clean_Output.csv')
print("\nSuccess! 10,000 lines processed completely. File 'Syntecxhub_10k_Clean_Output.csv' saved.")
```
