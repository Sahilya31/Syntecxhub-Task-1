```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ==========================================
# 1. Data Loading & Cleanup (10k records)
# ==========================================
np.random.seed(42)
N = 10000

# Generating dummy transactional data for the task
raw_data = {
    'InvoiceNo': np.random.randint(500000, 520000, size=N),
    'StockCode': np.random.randint(20000, 30000, size=N),
    'Quantity': np.random.randint(-10, 100, size=N),
    'InvoiceDate': pd.date_range(start='2025-01-01', end='2026-06-15', periods=N),
    'UnitPrice': np.random.uniform(1.0, 200.0, size=N),
    'CustomerID': np.random.choice([12345 + i for i in range(100)] + [np.nan], size=N, p=[0.0095]*100 + [0.05])
}
df = pd.DataFrame(raw_data)
print("Raw data shape:", df.shape)

# Cleaning process: removing missing IDs and cancelled orders (negative quantity)
df = df.dropna(subset=['CustomerID'])
df = df.drop_duplicates()
df = df[(df['Quantity'] > 0) & (df['UnitPrice'] > 0)]

# Final total spend column
df['TotalAmount'] = df['Quantity'] * df['UnitPrice']
print("Clean data shape:", df.shape)


# ==========================================
# 2. Calculating RFM Metrics
# ==========================================
# Target snapshot date (1 day after last invoice)
ref_date = df['InvoiceDate'].max() + pd.Timedelta(days=1)

# Grouping by customer to get individual RFM scores
rfm = df.groupby('CustomerID').agg({
    'InvoiceDate': lambda x: (ref_date - x.max()).days,
    'InvoiceNo': 'nunique',
    'TotalAmount': 'sum'
})

# Renaming for clarity
rfm.columns = ['Recency', 'Frequency', 'Monetary']


# ==========================================
# 3. Percentile Scoring & Segmentation
# ==========================================
# Ranking using method='first' to safely handle tie-breaks in qcut
rfm['R_score'] = pd.qcut(rfm['Recency'].rank(method='first'), q=4, labels=[4, 3, 2, 1])
rfm['F_score'] = pd.qcut(rfm['Frequency'].rank(method='first'), q=4, labels=[1, 2, 3, 4])
rfm['M_score'] = pd.qcut(rfm['Monetary'].rank(method='first'), q=4, labels=[1, 2, 3, 4])

# Creating the core combination string
rfm['RFM_Group'] = rfm['R_score'].astype(str) + rfm['F_score'].astype(str) + rfm['M_score'].astype(str)

# Custom mapping logic based on user groups
def assign_tag(customer):
    group = customer['RFM_Group']
    f = customer['F_score']
    m = customer['M_score']
    r = customer['R_score']

    if group == '444':
        return 'Champions'
    elif f == 4 and m == 4:
        return 'Loyal Customers'
    elif r == 4 and f == 1:
        return 'New Customers'
    elif r == 1:
        return 'Churn Risk'
    elif r == 2 and f <= 2:
        return 'About to Sleep'
    else:
        return 'Average / Potential'

rfm['Segment'] = rfm.apply(assign_tag, axis=1)
print("\nFinal Segment Counts:")
print(rfm['Segment'].value_counts())


# ==========================================
# 4. Plots & Visualizations
# ==========================================
sns.set_style("whitegrid")
counts = rfm['Segment'].value_counts()

# Plot 1: Pie Chart for Segment Share
plt.figure(figsize=(7, 7))
plt.pie(
    counts,
    labels=counts.index,
    autopct='%1.1f%%',
    startangle=140,
    colors=sns.color_palette('muted', len(counts)),
    wedgeprops={'edgecolor': 'white', 'linewidth': 1.5}
)
plt.title('Customer Segments Distribution (Pie Chart)', fontsize=14, pad=12)
plt.tight_layout()
plt.show()

# Plot 2: Bar Chart for Counts
plt.figure(figsize=(10, 5))
sns.barplot(x=counts.index, y=counts.values, palette='Set2', hue=counts.index, legend=False)
plt.title('Customer Count per Segment', fontsize=14, pad=12)
plt.xticks(rotation=20, ha='right')
plt.ylabel('Number of Customers')
plt.tight_layout()
plt.show()

# Plot 3: Metrics Spread (Histograms)
fig, axes = plt.subplots(1, 3, figsize=(16, 4))
fig.suptitle('Overall Distribution of RFM Metrics', fontsize=14, y=1.02)

sns.histplot(rfm['Recency'], kde=True, ax=axes[0], color='skyblue')
axes[0].set_title('Recency (Days)')

sns.histplot(rfm['Frequency'], kde=True, ax=axes[1], color='lightgreen')
axes[1].set_title('Frequency (Orders)')

sns.histplot(rfm['Monetary'], kde=True, ax=axes[2], color='salmon')
axes[2].set_title('Monetary (Total Spend)')

plt.tight_layout()
plt.show()

# Saving the processed dataframe to local system
rfm.to_csv('Syntecxhub_10k_Clean_Output.csv')
print("\nProcess finished! CSV saved successfully.")
```
